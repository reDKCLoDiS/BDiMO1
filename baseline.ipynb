```python
minimum_cor = []  # Создаем список для накопления результатов
for i in unidis.id2.unique():
    result = {}  # Для каждого i создаем отдельный словарь
    result['id1'] = i
    result['id2'] = unidis[(unidis.correlation == unidis[unidis.id2 == i].correlation.min()) & (unidis.id2 == i)].id1.values[0]
    result['corr'] = unidis[unidis.id2 == i].correlation.min()
    minimum_cor.append(result)  # Добавляем словарь в список
minimum_cor = pd.DataFrame(minimum_cor)
```

```python
preds = {'id': np.arange(10000, 15000, 1),
        'class': [np.nan] * len(np.arange(10000, 15000, 1)),
        'D_prob': [np.nan] * len(np.arange(10000, 15000, 1)),
        'K_prob': [np.nan] * len(np.arange(10000, 15000, 1))}
preds = pd.DataFrame(preds)
preds
```

```python
max_class = train.loc[max_cors.id2]['class'].values
min_class = train.loc[minimum_cor.id2]['class'].values
```

```python
max_cors['class'] = max_class
max_cors
```

```python
minimum_cor['class'] = min_class
minimum_cor = minimum_cor.replace({"K": "D", "D": "K"})
minimum_cor['corr'] = minimum_cor.apply(lambda x: abs(x['corr']), axis=1)
minimum_cor
```

```python
preds
```

```python
preds = preds.rename({'D_prob': 'D', 'K_prob': "K"}, axis=1)
preds
```

```python
max_cors = max_cors.rename({'corr': 'max_cors', 'class': 'max_class'}, axis=1)
minimum_cor = minimum_cor.rename({'corr': 'min_corr', 'class': 'min_class'}, axis=1)
```

```python
max_cors
```

```python
full_cors = pd.merge(max_cors, minimum_cor, how='outer', on=['id1', 'id2'])
class_prob = full_cors.drop(columns='id2')
```

```python
class_prob
```

```python
for i in preds.id:
    if max(class_prob[class_prob.id1 == i]['min_corr'].max(), class_prob[class_prob.id1 == i]['max_cors'].max()) == class_prob[class_prob.id1 == i]['max_cors'].max():
        print('max', class_prob.loc[class_prob[class_prob.id1 == i]['max_cors'].idxmax(), 'max_class'], class_prob[class_prob.id1 == i]['max_cors'].max())
        preds.loc[preds.id == i, class_prob.loc[class_prob[class_prob.id1 == i]['max_cors'].idxmax(), 'max_class']] = class_prob[class_prob.id1 == i]['max_cors'].max()
    else:
        print('min', class_prob.loc[class_prob[class_prob.id1 == i]['min_corr'].idxmax(), 'min_class'], class_prob[class_prob.id1 == i]['min_corr'].max())
        preds.loc[preds.id == i, class_prob.loc[class_prob[class_prob.id1 == i]['min_corr'].idxmax(), 'min_class']] = class_prob[class_prob.id1 == i]['min_corr'].max()
```

```python
preds
```

```python
# D < 0, K > 0
preds['class_prob'] = np.zeros(len(preds))
for i in preds.id:
    if max(preds[preds.id == i]['D'].max(), preds[preds.id == i]['K'].max()) == preds[preds.id == i]['D'].max():
        print('D', -preds[preds.id == i]['D'].max())
        preds.loc[preds.id == i, 'class_prob'] = -preds[preds.id == i]['D'].max()
    else:
        print('K', preds[preds.id == i]['K'].max())
        preds.loc[preds.id == i, 'class_prob'] = preds[preds.id == i]['K'].max()
```

```python
def sig(x):
 return 1/(1 + np.exp(-x))
predede = preds[['id', 'class', 'class_prob']] 
predede
```

```python
predede['class_prob'] = predede['class_prob'].apply(sig)
```

```python

predede['class'] = predede.class_prob.apply(lambda x: x > 1)
```
